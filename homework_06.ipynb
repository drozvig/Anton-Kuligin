Задание 1
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
The Moscow Times - Wednesday, October 2, 2002
The Guardian - Friday, 11.10.13
Daily News - Thursday, 18 August 1977


```python
from datetime import timedelta
from datetime import datetime

The_Moscow_Time = datetime.strptime("Wednesday, October 2, 2002", "%A, %B %d, %Y")#Создаем новый объект и форматируем строку в объект
The_Guardian = datetime.strptime("Friday, 11.10.13", "%A, %d.%m.%y")
Daily_News = datetime.strptime("Thursday, 18 August 1977", "%A, %d %B %Y") 

print(The_Moscow_Time)
print(The_Guardian)
print(Daily_News)
```

stream = [‘2018-04-02’, ‘2018-02-29’, ‘2018-19-02’]

Напишите функцию, которая проверяет эти даты на корректность. Т. е. для каждой даты возвращает True (дата корректна) или False (некорректная дата).


```python
stream = ['2018-04-02', '2018-02-29', '2018-19-02'] 

def сheckdate(dates):
    """
    Функция проверяет корректность дат
    """
    for date in dates:
        try:
            cdate = datetime.strptime(date, "%Y-%m-%d") 
        except:
            print(date, False)
        else:
            print(date, True)
сheckdate(stream)
```

Задание 3
Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date. 
Даты должны вводиться в формате YYYY-MM-DD. 
В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
def datelist(start_date,end_date):
    """
    Функция выводит список дат за указанный период
    """
    date_list = []
    try:
        sdate = datetime.strptime(start_date, "%Y-%m-%d") #начальная дата
        edate = datetime.strptime(end_date, "%Y-%m-%d") #конечная дата
    except:
        print(date_list) #в случаи если в вод не в формате печатаем пустой список
    else:
        while edate >= sdate: # ПОКА КОНЕЦ БОЛЬШЕ НАЧАЛА
            date_list.append(datetime.strftime(sdate, "%Y-%m-%d"))#Записываем в новый список дату при этом переводя из ДАТЕобъекта к строке
            sdate += timedelta(days=1) #Прибовляем к стартовой дате 1 день               
        print(date_list)
        
datelist('2021-01-01','2021-11-28')
```
