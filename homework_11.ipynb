Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
df = pd.read_csv('ratings.csv')
df.head()
```

```python
def rating_film(row):
    if float(row['rating']) <= 2:
        return 'Low'
    if float(row['rating']) <= 4:
        return 'Middle'
    return 'Highe'
```

```python
df_movie = pd.read_csv('movies.csv')
df_movie.head()
```

```python
df = df.groupby('movieId').agg({'rating':'mean'}).reset_index() #Нашли средний рэйтинг и сделали новый датафрэйм
df.head()
```

```python
df['class'] = df.apply(rating_film, axis=1)#добавили новый столбец использую функцию 
#df = df.drop('rating', 1)
df.head()
```

```python
df_movie = df_movie.merge(df, on = 'movieId')# Соединили два датафрэйма по Мувиайди
df_movie.tail()
```

Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
df_key = pd.read_csv('keywords.csv')
df_key.head()
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }
```

```python
def find_region(row):
    for i in geo_data.items():
        for a in i[1]:
            if a in row['keyword']:
                return  i[0]
    
    return 'undefined'
```

```python
df_key['region'] = df_key.apply(find_region, axis=1)
df_key.tail()
```
